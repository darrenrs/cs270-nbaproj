# Libraries and Parse Data

```python
import pandas as pd

def spread_df():
  data_df = pd.read_csv('combined_out.csv')
  data_df = data_df.drop('AwayTeam', axis=1)\
                   .drop('HomeTeam', axis=1)\
                   .drop('Date', axis=1)
  data_df = data_df.dropna(axis=0)

  return data_df

spread_df()
```
